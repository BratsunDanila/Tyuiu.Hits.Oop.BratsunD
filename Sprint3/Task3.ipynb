```
interface IFired
{
    public void Fired();
}

public class Employee
{
    private string name;
    private int salary;
    private int age;

    public int Age
    {
        get { return age; }
        set
        {
            if (value > 0)
                age = value;
            else
                throw new ArgumentOutOfRangeException("Возраст не может быть отрицательным!");
        }
    }

    public int Salary
    {
        get { return salary; }
        set
        {
            if (value > 0)
                salary = value;
            else
                throw new ArgumentOutOfRangeException("Зарплата не может быть отрицательной!");
        }
    }
    public string Name
    {
        get { return name; }
        set {name = value; }
    }

    public Employee(string name, int age, int salary)
    {
        Name = name;
        Age = age;
        Salary = salary;
    }

    public virtual void Fired(int cashback)
    {
        Console.WriteLine($"Сотрудник {Name} уволен, необходимо выплатить ему {cashback} рублей.");
    }


    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Зарплата: {Salary} рублей";
    }

    public virtual void Work()
    {
        Console.WriteLine($"Сотрудник {Name} работает");
    }

    public virtual void TakeVacation()
    {
        Console.WriteLine($"Сотрудник {Name} взял отпуск");
    }

    public virtual void TakeVacation(int date)
    {
        Console.WriteLine($"Сотрудник {Name} взял отпуск на {date} дней");
    }

     public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }

    public virtual void Compare(Employee otherEmployee)
    {
        int priceDifference = this.Salary - otherEmployee.Salary;
        if (Salary < otherEmployee.Salary)
        {
            Console.WriteLine ($"{Name} зарабатывает меньше, чем {otherEmployee.Name} на {-priceDifference} руб.");
        }
        else if (Salary > otherEmployee.Salary)
        {
            Console.WriteLine ($"{Name} зарабатывает больше, чем {otherEmployee.Name} на {priceDifference} руб.");
        }
        else
        {
            Console.WriteLine ($"{Name} и {otherEmployee.Name} имеют одинаковую зарплату.");
        }
    }

}

public class EmployeeList<T> where T : Employee
{
    private List<T> employees = new List<T>();

    public void AddEmployee(T employee)
    {
        employees.Add(employee);
        Console.WriteLine($"Сотрудник {employee.Name} добавлен в список.");
    }

    public void RemoveEmployee(T employee)
    {
        if (employees.Remove(employee))
        {
            Console.WriteLine($"Сотрудник {employee.Name} удален из списка.");
        }
        else
        {
            Console.WriteLine($"Сотрудник {employee.Name} не найден в списке.");
        }
    }

    public void DisplayAll()
    {
        foreach (var employee in employees)
        {
            employee.Display();
        }
    }

    public void FireEmployee(T employee)
    {
        RemoveEmployee(employee);
    }
}

public class Manager : Employee
{
    public int SubordinatesCount { get; set; }

    public Manager(string name, int age, int salary, int subordinatescount) : base (name, age, salary)
    {
        SubordinatesCount = subordinatescount;
    }

    public override void Work()
    {
        Console.WriteLine($"Менеджер {Name} управляет командой");
    }

    public override void Display()
    {
        Console.WriteLine($"Менеджер - {GetInfo()}, Количество подчиненных: {SubordinatesCount}");
    }

}

public class Developer : Employee
{
    public string Specialization { get; set; }

    public Developer(string name, int age, int salary, string specialization) : base (name, age, salary)
    {
        Specialization = specialization;
    }

    public override void TakeVacation()
    {
        Console.WriteLine($"Разработчик {Name} взял отпуск, его необходимо согласовать с проектными сроками!!!");
    }

    public override void Display()
    {
        Console.WriteLine($"Разработчик - {GetInfo()}, Специализация: {Specialization}");
    }

}

public class GameDeveloper : Developer, IFired
{

    public string Genre { get; set; }

    public GameDeveloper(string name, int age, int salary, string specialization, string genre) : base (name, age, salary, specialization)
    {
        Genre = genre;
    }

    public override void TakeVacation()
    {
        Console.WriteLine($"Разработчик игр {Name} взял отпуск, его необходимо согласовать с проектными сроками!!!");
    }

    public override void Display()
    {
        Console.WriteLine($"Разработчик игр - {GetInfo()}, Жанр игр: {Genre}");
    }

    public void Fired()
    {
        Console.WriteLine($"Разработчик игр {Name} УВОЛЕН!!!");
    }

}

public class Designer : Employee, IFired
{
    public string DesignStyle { get; set; }

    public Designer(string name, int age, int salary, string designstyle) : base (name, age, salary)
    {
        DesignStyle = designstyle;
    }

    public override void Work()
    {
        Console.WriteLine($"Дизайнер {Name} работает над дизайн-проектом №1");
    }

    public override void Display()
    {
        Console.WriteLine($"Дизайнер - {GetInfo()}, Стиль дизайна: {DesignStyle}");
    }

    public void Fired()
    {
        Console.WriteLine($"Дизайнер {Name} УВОЛЕН!!! Необходимо внести этого сотрудника в черный список дизайнеров");
    }

}

//Employee first = new Manager("Иван", 34, 250000, 30);
//Employee second = new Developer("Василий", 27, 120000, "Back-End");
//GameDeveloper forth = new GameDeveloper("Эдуард", 24, 80000, "", "Инди");
//Designer third = new Designer("Ксения", 25 , 150000, "Стиль Аниме");
//Employee forth = new Developer("Максим", -10 , -555, "Front-End");

//first.Display(); first.Work(); first.TakeVacation(90);
//second.Display(); second.TakeVacation();
//third.Display(); third.Work(); third.Fired(); third.Fired(50000);
//forth.Display(); forth.Fired();
//first.Compare(third);


EmployeeList<Employee> employeeList = new EmployeeList<Employee>();

Manager first = new Manager("Иван", 34, 250000, 30);
Developer second = new Developer("Василий", 27, 120000, "Back-End");
Designer third = new Designer("Ксения", 25 , 150000, "Стиль Аниме");
GameDeveloper forth = new GameDeveloper("Эдуард", 24, 80000, "", "Инди");

employeeList.AddEmployee(first);
employeeList.AddEmployee(second);
employeeList.AddEmployee(forth);
employeeList.AddEmployee(third);

Console.WriteLine("\nСписок сотрудников:");
employeeList.DisplayAll();

employeeList.FireEmployee(third);
employeeList.FireEmployee(third);

Console.WriteLine("\nОбновленный список сотрудников:");
employeeList.DisplayAll();

first.TakeVacation();
first.TakeVacation(50000);
forth.TakeVacation();
```
